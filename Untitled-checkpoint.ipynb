```python
import pandas as pd
import matplotlib.pyplot as plt

datos = pd.read_csv("products.csv")
df = datos[["productID", "categoryID", "unitPrice", "unitsInStock"]]

print(df.sample(5))

valor_por_ciudad = df.groupby("unitsInStock")["categoryID"].mean()
valor_por_ciudad.head(10).plot.barh()
plt.show()

df.unitPrice.value_counts().head(10).plot.pie()
plt.show()

df.unitPrice.plot.hist()
plt.show()
```
